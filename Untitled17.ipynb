```python
df = df.drop(['date', 'rating'], axis=1)
```

```python
#Get dummies in the variation column to have a batter score at model with a dataset more readable
var_dummies = pd.get_dummies(df['variation'])
var_dummies
```

```python
df.drop('variation', axis=1, inplace=True)
```

```python
df
```

```python
#Concat with dataset original
df = pd.concat([df, var_dummies], axis = 1)
df
```

```python
#Transform the verified_reviews column for binary values
from sklearn.feature_extraction.text import CountVectorizer
txt_data = df['verified_reviews']
#Using the stop words to make sure take just the features necessary
sample = CountVectorizer(stop_words='english')
df_vector = sample.fit_transform(txt_data)
```

```python
print(df_vector.toarray())
```

```python
df.drop('verified_reviews', axis=1, inplace=True)
```

```python
reviews = pd.DataFrame(df_vector.toarray())
```

```python
reviews
```

```python
df = pd.concat([df, reviews], axis=1)
df
```

```python
#Split the dataframe in train and test
x = df.drop('feedback', axis=1)
y = df['feedback']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
#Creatring the model with 1 input layer, 1 hidden layer and 1 output layer
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=400, activation='relu', input_shape= (3800,)))
model.add(tf.keras.layers.Dense(units=400, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
hist = model.fit(x_train, y_train, epochs=20)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
pred = model.predict(x_train)
```

```python
#Making sure the result is 0 or 1
pred = (pred > 0.5)
```

```python
#Look the score with confusion matrix
cm = confusion_matrix(y_train, pred)
cm
```

```python
#See based in heatmap
sns.heatmap(cm, annot=True);
```

```python
#See the score with test dataset
pred_test = model.predict(x_test)
pred_test = (pred_test > 0.5)
cm = confusion_matrix(y_test, pred_test)
cm
```

```python
sns.heatmap(cm, annot=True);
```

```python
plt.plot(hist.history['accuracy'])
plt.title('Model accuracy during training')
plt.xlabel('Epoch')
plt.ylabel('Training accuracy')
plt.legend('Trainig accuracy');
```
